```python
from datasets import load_dataset

#Loading our data
data = load_dataset("rotten_tomatoes")
data
```

```python
data["train"][0,-1] #First and last review
```

```python
import torch
print(torch.cuda.is_available())
print(torch.cuda.current_device())
```

```python
import numpy as np
from tqdm import tqdm
from transformers.pipelines.pt_utils import KeyDataset
```

# RoBERTa

This is a Encoder Architecture based Model (Representative Type)

```python
from transformers import pipeline

model_path = "cardiffnlp/twitter-roberta-base-sentiment-latest"

#load model into pipeline

pipe =(pipeline(model=model_path,tokenizer=model_path,return_all_scores=True,device="cuda:0"))
```

```python

#Run inference
y_pred=[]
for output in tqdm(pipe(KeyDataset(data["test"],"text")),total =len(data["test"])):
    #since model output has 3 scores --> negative,neutral, positive
    negative_score =output[0]["score"]
    positive_score =output[2]["score"]
    y_pred.append(np.argmax([negative_score,positive_score]))
```

```python
print(y_pred.count(0))
print(y_pred.count(1))
```

```python
#Evaluation

from sklearn.metrics import classification_report

print(classification_report(data["test"]["label"],y_pred,target_names=["negative","positive"]))

```

F1 score for both the classes are almost near which is good
From above we can see precision for negative class is less which means model  classifies positives as negative
and we can see recall for positive is less which means model classifies positives as negative


# Flan-T5

This is a Transformer Architecture Based Model(Generative Type)

```python
#prepare our data for prompting
prompt = "Is the following sentence positive or negative?"
data= data.map(lambda example: {"t5":prompt+example["text"]})
data
```

```python
#Loading th model
from transformers import pipeline
pipe = pipeline("text2text-generation",model="google/flan-t5-small",device="cuda:0")
```

```python
y_pred =[]
for output in tqdm(pipe(KeyDataset(data["test"],"t5")),total=len(data["test"])):
    text=output[0]["generated_text"]
    y_pred.append(0 if text=="negative" else 1)
```

```python
from sklearn.metrics import classification_report

print(classification_report(data["test"]["label"],y_pred,target_names=["negative","positive"]))

```
